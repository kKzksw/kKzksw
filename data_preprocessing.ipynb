### $\bullet$ Convert strings into numerotated categories

```python
df_raw.info()
```

```python
# find out the 'Object' variable
cat_columns = df_raw.select_dtypes(include=['object']).columns.tolist()
cat_columns
```

```python
# id  & customer_id & name & ssn &type_of_loan
print(df_raw[['id', 'customer_id','name','ssn','type_of_loan']])
```

```python
# id  & customer_id & name & ssn &type_of_loan are irrelevant variables,Drop them 
df_backup1 = df_raw.drop(['id','customer_id','name','ssn','type_of_loan'],axis=1)
len(df_raw.columns),len(df_backup1.columns)
```

```python
# month
print(df_backup1['month'])
print(df_backup1['month'].unique())
```

```python
# month
month_mapping = {
    'January': 0, 'February': 1, 'March': 2, 'April': 3, 'May': 4,
    'June': 5, 'July': 6, 'August': 7
}

df_backup1['month'] = df_backup1['month'].map(month_mapping)
print(df_backup1[['month']])
```

```python
# occupation
print(df_backup1['occupation'])
print(df_backup1['occupation'].unique())
```

```python
# occupation
occupation_mapping = {
    'Scientist': 0, 'Teacher': 1, 'Engineer': 2, 'Entrepreneur': 3, 'Developer': 4,
    'Lawyer': 5, 'Media_Manager': 6, 'Doctor': 7, 'Journalist': 8, 'Manager': 9, 
    'Accountant': 10, 'Musician': 11, 'Mechanic': 12, 'Writer': 13, 'Architect': 14
}

df_backup1['occupation'] = df_backup1['occupation'].map(occupation_mapping)
print(df_backup1['occupation'])
```

```python
# credit_mix
print(df_backup1['credit_mix'])
print(df_backup1['credit_mix'].unique())
```

```python
# occupation
credit_mix_mapping = {
    'Good': 0, 'Standard': 1, 'Bad': 2
}

df_backup1['credit_mix'] = df_backup1['credit_mix'].map(credit_mix_mapping)
print(df_backup1['credit_mix'])
```

```python
# payment_of_min_amount
print(df_backup1['payment_of_min_amount'])
print(df_backup1['payment_of_min_amount'].unique())
```

```python
# payment_of_min_amount
payment_of_min_amount_mapping = {
    'No': 0, 'Yes': 1
}

df_backup1['payment_of_min_amount'] = df_backup1['payment_of_min_amount'].map(payment_of_min_amount_mapping)
print(df_backup1['payment_of_min_amount'])
```

```python
# payment_behaviour
print(df_backup1['payment_behaviour'])
print(df_backup1['payment_behaviour'].unique())
```

```python
# payment_behaviour
payment_behaviour_mapping = {
    'Low_spent_Small_value_payments': 0, 'Low_spent_Medium_value_payments': 1,'Low_spent_Large_value_payments': 2,
    'High_spent_Small_value_payments': 3, 'High_spent_Medium_value_payments': 4,'High_spent_Large_value_payments': 5
}

df_backup1['payment_behaviour'] = df_backup1['payment_behaviour'].map(payment_behaviour_mapping)
print(df_backup1['payment_of_min_amount'])
```
